```python
import numpy as np

def isValid(A):
    x = A[-1]
    for i in x:
        if i < 0:
            return True
    return False

def pivote(A):
    i, j = np.inf, 0
    for t, n in enumerate(A[-1]):
        if n < 0 and n < i:
            j = t
            i = n

    l = np.inf
    i = 0

    for t, n in enumerate(A[:, -1]):
        if A[t][j] == 0:
            continue  # Ignora esta fila, ya que el elemento en la columna pivote es cero
        elif A[t][j] > 0:
            ratio = n / A[t][j]
            if ratio < l:
                l = ratio
                i = t

    return i, j

def display_tabla(A, num_vars, num_holguras, decimales=2):
    # Encabezado de las variables de decisión y de holgura
    encabezado = ""
    for i in range(num_vars):
        encabezado += f"\tx{i+1}"
    for i in range(num_holguras):
        encabezado += f"\ts{i+1}"
    
    encabezado += "\tr"  # Agregar la columna para el resultado
    print(encabezado)
    
    # Formato de decimales para los valores
    formato = f"{{:.{decimales}f}}"  # Formato dinámico para decimales

    # Mostrar los elementos de la matriz en formato tabular
    for i in range(A.shape[0]):  # Recorre cada fila
        fila = ""
        for j in range(A.shape[1]):  # Recorre cada columna
            fila += "\t" + formato.format(A[i, j])  # Aplicar el formato de decimales
        print(fila)

def met_simplex(c, A, b):
    I = np.identity(len(b))
    A = np.hstack((A, I))
    A = np.hstack((A, np.array(b).reshape(-1, 1)))

    c = c + [0] * (len(b) + 1)
    A = np.vstack((A, c))

    while isValid(A):
        i, j = pivote(A)
        print(f"Pivote en fila {i}, columna {j}")
        display_tabla(A, len(c) - len(b) - 1, len(b))
        piv = A[i][j]
        if piv == 0:
            print("Error: pivote igual a cero, no se puede continuar.")
            break

        A[i] = A[i] / piv  # Divide la fila pivote por el pivote para hacer que el pivote sea 1

        for k in range(len(A)):
            if k != i:
                A[k] = A[k] - A[k][j] * A[i]

    print("FINAL")
    display_tabla(A, len(c) - len(b) - 1, len(b))

```

```python
c = [-50, -40, -70, -60]

A = [
    [2, 1, 3, 4],  # Tiempo de trabajo
    [3, 2, 4, 2],  # Madera
    [1, 1, 2, 3]   # Espacio en almacén
]

b = [200, 300, 150]
```

```python
met_simplex(c,A,b)
```
